# Data import and Analysis

```python
import numpy as np
import pandas as pd

import seaborn as sns
sns.set_style('whitegrid')


import matplotlib.pyplot as plt
%matplotlib inline
plt.rc("axes.spines", bottom= False, top=False, right=False)

df = pd.read_csv('World Happiness Report 2022.csv')
df.head(10)
```

```python
df.tail()
```

```python
df.describe()
```

```python
print(df.shape, df.columns, sep= '\n\n')
print()
print(df.info())
```

```python
df.apply('nunique')
# 5 countries are missing Happiness Scores
```

```python
df['Happiness score'].describe()
```

```python
unique_scores = df['Happiness score'].unique()
df['Happiness score'].unique()
```

```python
dont_need = ['Whisker-high', 'Whisker-low']
```

```python
df = df.drop(dont_need, axis = 1)
```

```python
df.head()
```

```python
df = df.rename({"Happiness score" : "Happiness"}, axis = 1)
df = df.rename({"Dystopia (1.83) + residual" : "Dystopia"}, axis=1)
df = df.rename({"Explained by: Healthy life expectancy" : "Life Expectancy"}, axis=1)
df = df.rename({"Explained by: Social support" : "Social Support"}, axis =1)
df = df.rename({"Explained by: GDP per capita" : "Gross Domestic Product"}, axis=1)
df = df.rename({"Explained by: Freedom to make life choices" : "Freedom"}, axis=1)
df = df.rename({"Explained by: Generosity" : "Generosity"}, axis=1)
df = df.rename({"Explained by: Perceptions of corruption" : "Perception of Corruption"}, axis=1)
df.set_index("RANK",inplace=True)
```
